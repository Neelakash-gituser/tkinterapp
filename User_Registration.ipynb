```python
from tkinter import*
import sqlite3
try:
    db=sqlite3.connect('my_data_set1.db')
    cr=db.cursor()
    cr.execute("create table user_reg(First_Name,Last_Name, Middle_Name,Date_of_Birth,Gender,City,State,Contact,Project_Name)")    
    db.commit()
except:
    print("Table is already present")
    
w=Tk()
x=StringVar()
y=StringVar()
z=StringVar()
m=StringVar()
n=StringVar()
o=StringVar()
p=StringVar()
q=StringVar()
r=StringVar()

w.title('Users Registration form')
w.configure(background='cyan')
w.geometry('800x600')

def insert():
    a=x.get()
    b=y.get()
    c=z.get()
    d=m.get()
    e=n.get()
    f=o.get()
    g=p.get()
    h=q.get()
    i=r.get()
    cr.execute("insert into user_reg(First_Name,Last_Name, Middle_Name,Date_of_Birth,Gender,City,State,Contact,Project_Name)values(?,?,?,?,?,?,?,?,?)",(a,b,c,d,e,f,g,h,i))
    db.commit()
    print("Submitted Successfully !")
    

L0=Label(w,text="USER REGISTRATION PORTAL",font='arial 20')
L0.place(x=180,y=10)

L1=Label(w,text='First Name :',font='Mangal 12',fg='black')
L1.place(x=3,y=60)
E1=Entry(w,font='Mangal 10',textvar=x)
E1.place(x=140,y=65)

L2=Label(w,text='Last Name :',font='Mangal 12',fg='black')
L2.place(x=3,y=105)
E2=Entry(w,font='Mangal 10',textvar=y)
E2.place(x=140,y=110)

L3=Label(w,text='Middle Name :',font='Mangal 12',fg='black')
L3.place(x=400,y=65)
E3=Entry(w,font='Mangal 10',textvar=z)
E3.place(x=540,y=70)

L4=Label(w,text='Date of Birth :',font='Mangal 12',fg='black')
L4.place(x=3,y=150)
E4=Entry(w,font='Mangal 10',textvar=m)
E4.place(x=140,y=155)

L5=Label(w,text='Gender :',font='Mangal 12',fg='black')
L5.place(x=400,y=140)
E5=Radiobutton(w,text='Male',variable=n,value="Male")
E0=Radiobutton(w,text='Female',variable=n,value="Female")
E12=Radiobutton(w,text='Other',variable=n,value="Other")
E5.pack()
E0.pack()
E12.pack()
E5.place(x=500,y=145)
E0.place(x=560,y=145)
E12.place(x=632,y=145)

L6=Label(w,text='City :',font='Mangal 12',fg='black')
L6.place(x=3,y=195)
E6=Entry(w,font='Mangal 10',textvar=o)
E6.place(x=140,y=200)

L7=Label(w,text='State :',font='Mangal 12',fg='black')
L7.place(x=3,y=240)
drop=OptionMenu(w,p,"West Bengal","Assam")
drop.pack()
drop.place(x=140,y=245)

L8=Label(w,text='Contact :',font='Mangal 12',fg='black')
L8.place(x=3,y=290)
E8=Entry(w,font='Mangal 10',textvar=q)
E8.place(x=140,y=295)

L9=Label(w,text='Project Name :',font='Mangal 12',fg='black')
L9.place(x=3,y=335)
E9=Entry(w,font='Mangal 10',textvar=r)
E9.place(x=140,y=340)

B1=Button(w,text='Submit',padx=12,pady=8,font='bold 12',command=insert)
B1.place(x=140,y=500)

w.mainloop()
```
